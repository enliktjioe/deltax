```python
import cv2
vidcap = cv2.VideoCapture('20201026-152144224/20201026-152144224.mp4')
success,image = vidcap.read()
count = 0
output = 'frames2/'
while success:
  cv2.imwrite(output + "frame%d.png" % count, image)     # save frame as JPEG file      
  success,image = vidcap.read()
  print('Read a new frame: ', success)
  count += 1
```
