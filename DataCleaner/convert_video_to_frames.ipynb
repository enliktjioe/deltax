```python
import cv2

# file_name = '20201223-141426796'
input_folder = 'RCS_Recording/'
file_name = '20201223-140835981'

# newpath = 'C:\\Users\\handy89\\deltax\\RCSnail-AI-lite\\src\\preprocessed\\' + file_name + '\\images'
newpath = input_folder + file_name + '/images/'


if not os.path.exists(newpath):
    os.makedirs(newpath)

vidcap = cv2.VideoCapture(input_folder + file_name + '.m4v')
success,image = vidcap.read()
count = 0

print(success)

while success:
#   cv2.imwrite(newpath + "\\frame%d.jpg" % count, image)     # save frame as JPEG file      
  cv2.imwrite(newpath + "frame%d.jpg" % count, image)     # save frame as JPEG file      
  success,image = vidcap.read()
  print('Read a new frame: ', success)
  count += 1
```
